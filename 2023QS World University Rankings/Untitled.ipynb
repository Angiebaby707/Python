```python
# what is the distribution of rankings across all universities?
# How many universities are ranked in different ranges(e.g.., top 100,101- 500,-1000)
```

```python
# Distribution of Rankings
# plot the distributions of rankings
University['Rank'].plot(kind='hist',bins=20,title = 'Distribution of University Rankings')
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation between Metrics and Rank:
1. What is the correlation between Academic Reputation, Employer Reputation, Faculty-Student Ratio, and the overall ranking?
2. Which metric has the strongest correlation with university ranking?
```

```python
# Calculate correlation between ranking and key metrics
metrics = ['ar score', 'er score', 'fsr score']
correlation = University[['Rank'] + metrics].corr()

print("Correlation between Rank and Key Metrics.")
print(correlation)
```

```python
# Which countries have the most universities in the top 100, 500, and 1000?
# How do the rankings of universities compare across different countries?
top_100_countries = University[University['Rank'] <=100]['location'].value_counts()
print (top_100_countries)
plt.figure(figsize = (10,6))
top_100_countries.plot(kind = 'bar', color='red')
plt.title('Number in Top 100 by country')
plt.xlabel('Country')
plt.ylabel('Number of Universities')
plt.show()
```

```python

# 1. What are the average scores of Academic Reputation (ar score), Employer Reputation (er score),
# and Faculty-Student Ratio (fsr score) across the top 100 universities?
# 2.How do these scores compare with universities ranked below 500?
# calculate average scores for key metrics in the top 100 and below 500
top_100_metrics = University[University['Rank']<=100][['ar score', 'er score', 'fsr score']].mean()
print(top_100_metrics)
plt.figure(figsize=(8,8))
plt.pie(top_100_metrics,labels = top_100_metrics.index,autopct='%1.1f%%',startangle=90,colors=plt.cm.Paired.colors)
plt.title('Average Scores For Key Metrics In Top 100 Universities')
plt.axis('equal') # Ensures the pie chart is drawn as a circle
plt.tight_layout()
plt.show()
```

```python
below_500_metrics = University[University['Rank'] > 500][['ar score','er score','fsr score']].mean()
print(below_500_metrics)
```
